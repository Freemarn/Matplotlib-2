## SUBPLOTS

```python
## This is done when you try to plot too many graphs on one figure
```

```python
import numpy as np
x = np.linspace(-20,20,100)
```

```python
sine = np.sin(x)
cos = np.cos(x)
plt.plot(x, sine)
plt.plot(x, cos)
```

```python
plt.subplots(2)
```

```python
fig1, ax= plt.subplots(2)
```

```python
ax
```

```python
ax[0].plot(x, sine)
ax[1].plot(x, cos)
fig1
```

```python
# plt.subplots(figsize = (10, 10))
fig1, ax= plt.subplots(1,2,figsize = (15, 5))
ax[0].plot(x, sine)
ax[1].plot(x, cos)
ax[0].set_title("Sine Graph")
ax[1].set_title("Cos Graph")
# fig1(figsize(25,25))
```
